```
import pandas as pd
import scipy
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```
home_table = pd.read_csv('/content/drive/My Drive/Data Analytics/FPL/538 data 22.10.2020/FBref test files/PL Home.csv', encoding = "ISO-8859-1", engine='python')
away_table = pd.read_csv('/content/drive/My Drive/Data Analytics/FPL/538 data 22.10.2020/FBref test files/PL Away.csv', encoding = "ISO-8859-1", engine='python')
```

```
home_table.rename(columns={'ï»¿team': 'team'}, inplace=True)
home_table = home_table.replace('Wolverhampton Wanderers','Wolverhampton')
home_table = home_table.replace('Leeds United', 'Leeds Utd')
home_table = home_table.replace('West Ham', 'West Ham Utd')
home_table['H Avg GS'] = home_table ['GF'] / home_table ['MP']
home_table ['H Avg GA'] = home_table ['GA'] / home_table ['MP']
home_table ['LA GS'] = home_table['H Avg GS'].mean()
home_table ['LA GA'] = home_table['H Avg GA'].mean()
home_table ['H AS'] = home_table['H Avg GS']/home_table ['LA GS']
home_table ['H DS'] = home_table['H Avg GA']/home_table ['LA GA']
home_table
```

```
away_table.rename(columns={'ï»¿Squad': 'team'}, inplace=True)
away_table = away_table.replace('Wolverhampton Wanderers','Wolverhampton')
away_table = away_table.replace('Leeds United', 'Leeds Utd')
away_table = away_table.replace('West Ham', 'West Ham Utd')
away_table['H Avg GS'] = away_table ['GF'] / away_table ['MP']
away_table ['H Avg GA'] = away_table ['GA'] / away_table ['MP']
away_table ['LA GS'] = away_table['H Avg GS'].mean()
away_table ['LA GA'] = away_table['H Avg GA'].mean()
away_table ['A AS'] = away_table['H Avg GS']/away_table ['LA GS']
away_table ['A DS'] = away_table['H Avg GA']/away_table ['LA GA']
away_table
```

```
home_table_reduced = home_table[['team', 'H AS', 'H DS']]
away_table_reduced = away_table[['team', 'A AS', 'A DS']]
combined_table = pd.merge(home_table_reduced, away_table_reduced, on='team')
combined_table.sort_values('team')
combined_table
```
